```python
file_length = 10000*np.array([1,1,1,1,1])
```

```python
averaged_detectionrate = [ddf.successful_reconstruction.sum()/10 for ddf in ddf_group_by_z_list] # divide by number of z 
```

```python
false_detection_number = []
for i in range(n_files):
    false_detection_number.append(ddf_group_by_reconz_list[i].false_reconstruction.sum()/file_length[i]) # divide by total number of files
```

```python
false_detection_rate = [np.array([])]*n_files
for i in range(n_files):
    false_detection_rate[i] = np.zeros((10,int(file_length[i]/100)))
    for j in range(10): # number of a/c
        for k in range(int(file_length[i]/100)): # number of halo per 
            try: 
                false_detection_rate[i][j,k] = false_detection_number[i][j][k]
            except:
                false_detection_rate[i][j,k] = 0
```

```python
detection_rate = [np.array([])]*n_files
for i in range(n_files):
    detection_rate[i] = np.zeros((10,int(file_length[i]/100)))
    for j in range(10): # number of a/c
        for k in range(int(file_length[i]/100)):
            try: 
                detection_rate[i][j,k] = averaged_detectionrate[i][j][k]
            except:
                detection_rate[i][j,k] = 0
```

```python
detection_rate_std = np.zeros((n_files,10))
detection_rate_mean = np.zeros((n_files,10))
for i in range(n_files):
    for j in range(10): # number of z
        detection_rate_std[i,j] = np.std(detection_rate[i][j])/(np.sqrt(file_length[i]/100))
        detection_rate_mean[i,j] = np.mean(detection_rate[i][j])
false_detection_rate_std = np.zeros((n_files,10))
false_detection_rate_mean = np.zeros((n_files,10))
for i in range(n_files):
    for j in range(10):
        false_detection_rate_std[i,j] = np.std(false_detection_rate[i][j])/(np.sqrt(file_length[i])) # std divided by sqrt of sample nu
        false_detection_rate_mean[i,j] = np.sum(false_detection_rate[i][j])
```

Expected false detection should be 



number of false detection given some halo in the dictionary * number of dictionary in full sky * probability of having that halo in the dictionary

```python
import pyccl as ccl
def halo_density(log_m):
    h = 67.4/100
    Omega_c = 0.12/h**2
    Omega_b = 0.0224/h**2
    cosmo = ccl.Cosmology(Omega_c=Omega_c, Omega_b=Omega_b,
                          h=0.674, sigma8=0.811, n_s=0.965, Neff=2.99) #placnk 2018
    M_vir = 10 ** log_m
    conc = 4
    hmfunc = []
    for z in input_redshift:
        halo1    =  hmod.triaxialJS02(mass=M_vir,conc=conc,redshift=z,ra=0.,dec=0,a_over_c=1,a_over_b=1, tri_nfw = False, long_truncation = True, OLS03 = True) #33,16
        hmd = ccl.halos.MassDef(halo1.Omega_z*halo1.Delta_vir, 'critical')
        hmf = ccl.halos.MassFuncTinker10(cosmo, mass_def = hmd) # boquet16 is "not compatible"
        a = 1 / (1 + z-0.0225)
        a2 = 1/(1+z+0.0225)
        dlogm = 0.2
        volume = 4/3*np.pi*ccl.comoving_radial_distance(cosmo, a2)**3 - 4/3*np.pi*ccl.comoving_radial_distance(cosmo, a)**3
        this_hmf = hmf.get_mass_function(cosmo, M_vir, a)*volume*dlogm 
        #/(4*np.pi*((180/np.pi))**2). This would convert to per degree^2
        hmfunc.append(this_hmf)
        #print(z)
    return hmfunc

halo_density_array = np.zeros((n_files,10))
for i in range(n_files):
    halo_density_array[i] = halo_density(true_log_mass[i])
```

```python
import pyccl as ccl
def halo_density(log_m):
    h = 67.4/100
    Omega_c = 0.12/h**2
    Omega_b = 0.0224/h**2
    cosmo = ccl.Cosmology(Omega_c=Omega_c, Omega_b=Omega_b,
                          h=0.674, sigma8=0.811, n_s=0.965, Neff=2.99) #placnk 2018
    M_vir = 10 ** log_m
    conc = 4
    hmfunc = []
    for z in input_redshift:
        halo1    =  hmod.triaxialJS02(mass=M_vir,conc=conc,redshift=z,ra=0.,dec=0,a_over_c=1,a_over_b=1, tri_nfw = False, long_truncation = True, OLS03 = True) #33,16
        hmd = ccl.halos.MassDef(halo1.Omega_z*halo1.Delta_vir, 'critical')
        hmf = ccl.halos.MassFuncTinker10(cosmo, mass_def = hmd) # boquet16 is "not compatible"
        a = 1 / (1 + z-0.0225)
        a2 = 1/(1+z+0.0225)
        dlogm = 0.2
        volume = 4/3*np.pi*ccl.comoving_radial_distance(cosmo, a2)**3 - 4/3*np.pi*ccl.comoving_radial_distance(cosmo, a)**3
        this_hmf = hmf.get_mass_function(cosmo, M_vir/0.674, a)*volume*dlogm 
        #/(4*np.pi*((180/np.pi))**2). This would convert to per degree^2
        hmfunc.append(this_hmf)
        #print(z)
    return hmfunc
```
