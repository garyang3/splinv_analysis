```python
%reload_ext autoreload
%autoreload 2
import numpy as np
from splinv import detect
from splinv import hmod
import matplotlib.pyplot as plt
```

```python
z_h1     =  0.2875
z_h2     =  0.2875
log_m1   =  14.8
log_m2   =  14.8
M_200_1   =  10.**(log_m1)
M_200_2  = 10.**(log_m2)
conc    =  4.
halo1    =  hmod.triaxialJS02(mass=M_200_1,conc=conc,redshift=z_h1,ra=0.,dec=0,a_over_c=1,a_over_b=1, tri_nfw = True, long_truncation = True, OLS03 = False) #33,16
halo2    =  hmod.nfwWB00(mass=M_200_1,conc=conc,redshift=z_h1,ra=0.,dec=0, long_truncation = True)
```

```python
ra, dec = np.linspace(50,5000,1000), np.linspace(50,100,1000)
dsigma1 = halo1.DeltaSigmaComplex(ra,dec)
dsigma2 = halo2.DeltaSigmaComplex(ra,dec)

sigma1 = halo1.Sigma(ra,dec)
sigma2 = halo2.Sigma(ra,dec)
```

```python
plt.plot(ra, sigma1.real,label='gary')
plt.plot(ra, sigma2.real,label='WB00')
plt.legend()
plt.show()
```

```python
plt.plot(ra, dsigma1.real,label='gary')
plt.plot(ra, dsigma2.real,label='WB00')
plt.legend()
plt.show()
```
