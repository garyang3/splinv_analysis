```python
%pylab inline
import numpy as np
from splinv import detect
from splinv import hmod
from splinv import darkmapper
from splinv.grid import Cartesian
from configparser import ConfigParser
import splinv
splinv.hmod_v2
import time
import pyfftw
import matplotlib.pyplot as plt
```

```python
# halo simulation
z_h     =  0.2425
log_m   =  14.745
M_200   =  10.**(log_m)
conc    =  4.
halo    =  hmod.triaxialJS02(mass=M_200,conc=conc,redshift=z_h,ra=0.,dec=0.,a_over_b = 1, a_over_c = 1)
```

```python
ra, dec = np.meshgrid(np.linspace(-100,100,100),np.linspace(-100,100,100))
```

```python
kmap = halo.Sigma(ra,dec)
```

```python
ks2D = hmod.ksmap(100,100)
```

```python
gmap = ks2D.transform(kmap, inFou = False, outFou = False)
```

```python
plt.imshow(gmap.imag)
```

```python
gmap_fftw = ks2D.transform_fftw(kmap, inFou = False, outFou = False)
```

```python
plt.imshow(gmap_fftw.imag)
```
